```
import numpy as np
import pandas as pd
```

```
data=pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name="Data")
```

```
data. isnull().sum()
```

```
data1=data. drop(columns=["ID",'ZIP Code'],axis=1)
data1.head()
```

```
data1. corr()
```

```
import statsmodels.api as sm
```

```
x=data1[["Age","Experience", "Income",
"Family","CCAvg","Education","Mortgage",
"Securities Account","CD Account","Online","CreditCard"]]
y=data1["Personal Loan"]
x1=sm.add_constant(x)
logistic=sm.Logit(y,x1)
result=logistic.fit()
result.summary()
```

Since Pvalue for Income,Family,CCAvg,Education,Securities Account,CD account,Online,Credit card is less than 0.05.
Therefore these are significant, while Income, Family,Education,CD account,Online,Credit card being the most significant(P-VALUE=0)
