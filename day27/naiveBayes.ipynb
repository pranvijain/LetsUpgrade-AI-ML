```
import pandas as pd
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```
data=pd.read_csv("train-1.csv")
features=[ 'Survived', 'Pclass', 'Sex', 'SibSp',
       'Parch', 'Embarked']
data.drop(["PassengerId","Name","Ticket","Fare","Cabin","Age"], axis=1,inplace=True)
data.isnull().sum()
```

```
label_encoder=preprocessing.LabelEncoder()
label_encoder.fit(data["Sex"])
data["Sex"]=label_encoder.transform(data["Sex"])
label_encoder.fit(data["Embarked"])
data["Embarked"]=label_encoder.transform(data["Embarked"])
data.head()
```

```
for feat in features:
  y=data[feat]
  x=data.drop(feat,axis=1)
  x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.25)
  clf=GaussianNB()
  y_pred=clf.fit(x_train,y_train).predict(x_test)
  print("Accuracy Score for {a} is {b}".format(a=feat,b=accuracy_score(y_test,y_pred,normalize=True)))
  print("Confusion Matrix for {a} is \n {b}".format(a=feat, b= confusion_matrix(y_test,y_pred)))
  print("\n")
```
